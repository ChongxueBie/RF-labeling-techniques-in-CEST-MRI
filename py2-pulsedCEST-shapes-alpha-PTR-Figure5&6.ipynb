```python
def raisedRecDesign(span, L):
# """
# Raised cosine FIR filter design
# Parameters:
# alpha : roll-off factor
# span : filter span in symbols
# L : oversampling factor (i.e, each symbol contains L samples)
# Returns:
# p - filter coefficients b of the designed
# FIR raised cosine filter
# """
#     print alpha
#     print span
#     print L
    t = np.arange(-span/2, span/2, 1/L) # +/- discrete-time base
    num = len(t)
    p = ones((1,num))*0.5
    p = squeeze(p)

    return p
```

```python
# def raisedCosineDemo():
# # """
# # Raised Cosine pulses and their manifestation in frequency domain
# # Usage:
# # >> from DigiCommPy.pulseshapers import raisedCosineDemo
# # >> raisedCosineDemo()
# # """
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftshift

# Tsym = 0.1 # Symbol duration in seconds
L = 100. # oversampling rate, each symbol contains L samples
span = 1. # filter span in symbols
Tsym= [200/1000.] # RC roll-off factors - valid range 0 to 1
# Fs = L/Tsym # sampling frequency
lineColors = ['b','r','g','k']
fig, (ax1,ax2) = plt.subplots(1, 2)
    
for i, Tsym_use in enumerate(Tsym):
    Fs = L/Tsym_use
    b = raisedRecDesign(span,L) # RC Pulse design
    
    # time base for symbol duration
    t = Tsym_use* np.arange(-span/2, span/2, 1/L)
    
    # plot time domain view
    ax1.plot(t,b,lineColors[i])
#     xlim(-10,10)

    #Compute FFT and plot double sided frequency domain view
    NFFT = 1<<(len(b)-1).bit_length() #Set FFT length = nextpower2(len(b))
    vals = fftshift(fft(b,NFFT))
    freqs = Fs* np.arange(-NFFT/2,NFFT/2)/NFFT
    ax2.plot(freqs,abs(vals)/abs(vals[len(vals)//2]),lineColors[i])
#     xlim(-2,2)
    
ax1.set_title('Raised cosine pulse');
ax2.set_title('Frequency response')
ax1.legend();ax2.legend()
fig.show()
```
